Comece criando uma instância do preprocessador e definindo um texto exemplo que você deseja processar. Em seguida, utilize o método preprocess_text para realizar o preprocessamento do texto, exibindo, por fim, o resultado do texto processado para verificação e análise.

## **Avaliação de Modelos de Machine Learning**

### Passo 2: Importando a Classe
---

O módulo ModelEvaluator da biblioteca MedRoom.AI.DataEngineer oferece uma suite abrangente de métricas para avaliar de maneira eficaz os modelos de Machine Learning. Abaixo, vamos explorar um exemplo prático de como utilizar este módulo para avaliar um modelo hipotético.

```python
from medroom.dna.processors.utils.evalmetrics import ModelEvaluator
```

### Passo 3: Treinando o Modelo
---

Neste tutorial, iremos explorar um exemplo prático, utilizando o dataset Iris e um modelo de Random Forest, para demonstrar como empregar a classe ModelEvaluator da biblioteca MedRoom.AI.DataEngineer para avaliar de maneira eficiente os modelos de Machine Learning.

Inicialmente, vamos carregar o dataset Iris, dividir os dados em conjuntos de treino e teste, e treinar um modelo de Random Forest:

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar e treinar um modelo de classificação (Random Forest)
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Fazer previsões com o modelo treinado
predictions = clf.predict(X_test)
```

### Passo 4: Uso da Classe

Crie uma instância do avaliador:

```python
evalmetrics = ModelEvaluator()
```

#### Acurácia
---

Mede a fração de acertos do modelo em relação ao total de instâncias.

```python
# Testar a acurácia
accuracy = evalmetrics.accuracy(y_test, predictions)
print(f"Acurácia calculada: {accuracy:.2f}")
```

#### Relatório de Classificação
---

Fornece métricas detalhadas como precisão, recall e f1-score para cada classe.

```python
# Gerar e exibir o relatório de classificação como imagem
classification_report_img, report = evalmetrics.classification_report_img_generator(y_test, predictions)
```

```python
# Ver imagem
from IPython.display import display, Image
display(Image(classification_report_img.getvalue()))
```

```python
# Relatório de classificação em um dicionário único
flat_report = evalmetrics.flatten_classification_report(report)
print("Relatório de Classificação Achatado:", flat_report)
```

#### ROC-AUC
---

Calcula a área sob a curva ROC (Receiver Operating Characteristic). É uma métrica útil para avaliação da capacidade do modelo em distinguir entre classes.

```python
roc_auc = evalmetrics.roc_auc(y_test, predictions)
print(f"ROC AUC calculado: {roc_auc:.2f}")
```

#### Matriz de Confusão
----

Apresenta uma visualização das predições do modelo versus os valores reais, mostrando onde o modelo acertou e onde errou.

```python
# Gerar e exibir a matriz de confusão como imagem
confusion_matrix_img, normalized_confusion_matrix = evalmetrics.confusion_matrix_img_generator(y_test, predictions)
```

```python
# Plotar imagem
display(Image(confusion_matrix_img.getvalue()))
```
